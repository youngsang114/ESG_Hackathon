```python
import pandas as pd
```

## 12-13 주가데이터 

```python
# 12-13 업종 데이터, 15년 주가데이터, 16년 주가데이터 불러오기
df_1213 = pd.read_csv('./df_add_job/df_1213_add_job.csv', encoding= 'utf-8', index_col='종목명')
mv_2012 = pd.read_csv('./marketvalue/mv2012.csv', encoding= 'euc-kr', index_col='종목명')
mv_2013 = pd.read_csv('./marketvalue/mv2013.csv', encoding= 'euc-kr', index_col='종목명')
```

```python
# 주가데이터 종가만 가져오기 
mv_2012 = pd.DataFrame(mv_2012['종가'])
mv_2013 = pd.DataFrame(mv_2013['종가'])
```

```python
# 각 리스트 저장
list_1213 = df_1213.index
list_mv2012 = mv_2012.index
list_mv2013 = mv_2013.index
```

```python
# 12-13 업종데이터와 12년 주가데이터 공통된 리스트 출력
set1=set(list_1213)
set2=set(list_mv2012)

common_values = list(set1 & set2)
print(common_values)
len(common_values)
```

```python
# 공통 리스트 기반 12년 주가데이터 출력
mv_2012 = mv_2012.loc[common_values]
mv_2012.sort_index(inplace=True)
```

```python
# 12-13 업종데이터와 13년 주가데이터 공통된 리스트 출력
set1=set(list_1213)
set2=set(list_mv2013)

common_values = list(set1 & set2)
print(common_values)
len(common_values)
```

```python
# 공통 리스트 기반 13년 주가데이터 출력
mv_2013 = mv_2013.loc[common_values]
mv_2013.sort_index(inplace=True)
```

```python
mv_2013
```

```python
# 공통 리스트 기반 12년 주가데이터와 13년 주가데이터 병합
mv_1213 = pd.merge(mv_2012, mv_2013, on = '종목명' , how = 'left')
mv_1213.columns = ['2012_종가', '2013_종가']
mv_1213
```

```python
mv_1213.to_csv('mv_1213.csv')
```
