```python
import pandas as pd
```

```python
df_1112=pd.read_csv('../Winsorization/99% 이상치제거_original_1213.csv',index_col=0)[['업종명','종목명','Diff_E','Diff_S','Diff_G','등락률','연도']]
df_1213=pd.read_csv('../Winsorization/99% 이상치제거_original_1213.csv',index_col=0)[['업종명','종목명','Diff_E','Diff_S','Diff_G','등락률','연도']]
df_1314=pd.read_csv('../Winsorization/99% 이상치제거_original_1314.csv',index_col=0)[['업종명','종목명','Diff_E','Diff_S','Diff_G','등락률','연도']]
df_1415=pd.read_csv('../Winsorization/99% 이상치제거_original_1415.csv',index_col=0)[['업종명','종목명','Diff_E','Diff_S','Diff_G','등락률','연도']]
df_1516=pd.read_csv('../Winsorization/99% 이상치제거_original_1516.csv',index_col=0)[['업종명','종목명','Diff_E','Diff_S','Diff_G','등락률','연도']]
df_1617=pd.read_csv('../Winsorization/99% 이상치제거_original_1617.csv',index_col=0)[['업종명','종목명','Diff_E','Diff_S','Diff_G','등락률','연도']]
df_1718=pd.read_csv('../Winsorization/99% 이상치제거_original_1718.csv',index_col=0)[['업종명','종목명','Diff_E','Diff_S','Diff_G','등락률','연도']]

df=pd.concat([df_1112,df_1213,df_1314,df_1415,df_1516,df_1617,df_1718],axis=0).dropna(axis=0)
df
```

```python
X=df[['Diff_E','Diff_S','Diff_G']]
y=df['등락률']
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np

X_train,X_test,y_train,y_test= train_test_split(X.values,y,test_size=0.2, random_state=0)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_preds=lr.predict(X_test)
mse= mean_squared_error(y_test,y_preds)
rmse=np.sqrt(mse)

print(f"MSE : {mse:.3f}, RMSE:{rmse:.3f}")
print(f"Variance score:{r2_score(y_test,y_preds):.3f}")
```
