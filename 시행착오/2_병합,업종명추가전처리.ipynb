```python
import pandas as pd
import numpy as np
```

## 산업군까지 나와있는 csv파일의 업종명을 리스트로 만들기 
- 아까 만들어둔 csv파일에 종목별 분류를 추가하기
- 대분류로 바꿔줄 소분류들이다

- 종목분류가 나와있는 데이터를 kospi라는 데이터프레임에 저장
- kospi_list라고 종목명값들을 리스트로 저장

```python
kospi=pd.read_csv('./ESG_csv/코스피업종분류.csv',encoding='euc-kr',index_col='종목명')
kospi_list=kospi.index.tolist()
kospi_list
```

## 로직!!!
- 만든 csv파일을 각각 데이터프레임으로 만들고 필요없는 열인 ('ESG 통합등급','평가연도')을 삭제한다
- 두 연속적인 연도의 데이터프레임을 '종목명'을 가지고 inner로 merge를 한다
- 두 연도의 E,S,G의 차이인 열을 각각 만들어 준다
- '종목명'으로 인덱스를 바꿔준다
- 이렇게 만든 데이터프레임의 index('종목명')들의 리스트를 만들어준다
- set을 이용해 두 연도의 리스트안의 공통 값들을 common_values로 저장한다
- loc 필터를 이용해 kospi데이터프레임의 common_values('종목명')들을 데이터프레임형식으로 가져온다
- duplicated()를 이용해 중복치를 지워주고 sort_index()를 통해 종목명들의 오름차순으로 정렬한 후 list1이라는 데이터프레임에 저장한다
- 같은 방법으로 merge한 데이터프레임도 list2라는 데이터프레임에 저장한다 
- 이 과정을 통해 list1,list2는 같은 index의 개수와, 정렬을 같는다
- concat을 이용해 합쳐준다

### Diff 12/13
- 688개에서 507개로 row값이 줄었다

```python
df_12=pd.read_csv('./ESG_csv/2012.csv',encoding='euc-kr').drop(['ESG 통합등급','평가연도'] ,axis=1)
df_13=pd.read_csv('./ESG_csv/2013.csv',encoding='euc-kr').drop(['ESG 통합등급','평가연도'] ,axis=1)
```

```python
diff_1213=pd.merge(df_12,df_13,how='inner',on='종목명')
diff_1213['Diff_E']=''
diff_1213['Diff_S']=''
diff_1213['Diff_G']=''
diff_1213.set_index('종목명',drop=True,inplace=True)
diff_1213
```

```python
## name_1213은 diff_1213의 종목명을 리스트로 만든 것이다
## 688개에서 507개로 데이터가 줄었다

name_1213=diff_1213.index.to_list()
```

```python
set1=set(kospi_list)
set2=set(name_1213)

common_values = list(set1 & set2)
print(common_values)
len(common_values)
```

```python
### common_value는 산업군까지 나와있는 csv파일의 업종명을 리스트와 diff_1213년도이름 리스트에 있는 같은 종목명을 리스트로 만든 것이다

kospi.loc[common_values].sort_index()[['업종명']]
```

```python
list1=diff_1213.loc[common_values]
list1=list1.loc[~list1.index.duplicated()]
list1.sort_index(inplace=True)
```

```python
list2=kospi.loc[common_values].sort_index()[['업종명']]
list2
```

```python
df_1213_add_job=pd.concat([list1,list2],axis=1)
df_1213_add_job
#df_1213_add_job.to_csv('df_1213_add_job.csv')
```

### Diff 13/14
- 681개에서 507개로 row값이 줄었다

```python
df_13=pd.read_csv('./ESG_csv/2013.csv',encoding='euc-kr').drop(['ESG 통합등급','평가연도'] ,axis=1)
df_14=pd.read_csv('./ESG_csv/2014.csv',encoding='euc-kr').drop(['ESG 통합등급','평가연도'] ,axis=1)

diff_1314=pd.merge(df_13,df_14,how='inner',on='종목명')
diff_1314['Diff_E']=''
diff_1314['Diff_S']=''
diff_1314['Diff_G']=''
diff_1314.set_index('종목명',drop=True,inplace=True)
print(diff_1314.shape)

name_1314=diff_1314.index.to_list()

set1=set(kospi_list)
set2=set(name_1314)

common_values = list(set1 & set2)
#print(common_values)
#len(common_values)

list1=diff_1314.loc[common_values]
list1=list1.loc[~list1.index.duplicated()]
list1.sort_index(inplace=True)

list2=kospi.loc[common_values].sort_index()[['업종명']]

df_1314_add_job=pd.concat([list1,list2],axis=1)
df_1314_add_job
```
