## Origin 1213데이터 만들기

```python
list_1=pd.read_csv('./df_add_job/df_1213_add_job.csv')
```

```python
list_2=pd.read_csv('./MV_sum/mv_1213.csv')
```

```python
Sum_list=pd.merge(list_1,list_2,how='inner',on='종목명')
Sum_list
```

```python
## 12,13 E의 변화('object' --> 'float')
Sum_list['2012_E']=Sum_list['2012_E'].map({
    'A+': 5,
    'A' : 4,
    'B+': 3,
    'B' : 2,
    'B이하' : 1.5,
    'C':1
})
Sum_list['2013_E']=Sum_list['2013_E'].map({
    'A+': 5,
    'A' : 4,
    'B+': 3,
    'B' : 2,
    'B이하' : 1.5,
    'C':1
})

## 12,13 S의 변화('object' --> 'float')
Sum_list['2012_S']=Sum_list['2012_S'].map({
    'A+': 5,
    'A' : 4,
    'B+': 3,
    'B' : 2,
    'B이하' : 1.5,
    'C':1
})
Sum_list['2013_S']=Sum_list['2013_S'].map({
    'A+': 5,
    'A' : 4,
    'B+': 3,
    'B' : 2,
    'B이하' : 1.5,
    'C':1
})

## 12,13 G의 변화('object' --> 'float')
Sum_list['2012_G']=Sum_list['2012_G'].map({
    'A+': 5,
    'A' : 4,
    'B+': 3,
    'B' : 2,
    'B이하' : 1.5,
    'C':1
})
Sum_list['2013_G']=Sum_list['2013_G'].map({
    'A+': 5,
    'A' : 4,
    'B+': 3,
    'B' : 2,
    'B이하' : 1.5,
    'C':1
})

## Diff값 채우기
Sum_list['Diff_E']= Sum_list['2013_E']-Sum_list['2012_E']
Sum_list['Diff_S']= Sum_list['2013_S']-Sum_list['2012_S']
Sum_list['Diff_G']= Sum_list['2013_G']-Sum_list['2012_G']
Sum_list['등락률']= (Sum_list['2013_종가']-Sum_list['2012_종가'])/ Sum_list['2012_종가']
Sum_list['연도']='12,13년'
```
