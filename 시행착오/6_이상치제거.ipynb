```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
origin_1213 = pd.read_csv('./Origin_Data/original_1213.csv')

# 데이터 예시
data = origin_1213['등락률']

# 이상치 임계값 설정 (98%)
threshold = np.percentile(data, 98)

# 박스 플롯 그리기
plt.boxplot(data)

# 이상치 표시
outliers = [val for val in data if (val > threshold) |(val <-threshold)]
for outlier in outliers:
    plt.plot(1, outlier, 'ro')  # 이상치를 빨간색으로 표시

# 그래프 보여주기
plt.show()

outliers = [val for val in data if (val > threshold)|(val <-threshold)]
print("이상치 데이터:", outliers)
len(outliers)

# 이상치 제외한 값을 원본에 저장
origin_1213 = origin_1213[~origin_1213['등락률'].isin(outliers)]

#origin_1213.to_csv('98% 이상치제거_original_1213.csv')
origin_1213
```
