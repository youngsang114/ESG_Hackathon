> 위 과정 반복

```python
def transform(df):
    
    df.drop('Unnamed: 0', axis=1, inplace=True)
    
    df['업종명'] = df['업종명'].map({
        '건설업': '건설업',
        '기타금융' : '금융 및 보험업',
        '증권': '금융 및 보험업',
        '보험' : '금융 및 보험업',
        '은행' : '금융 및 보험업',
        '서비스업' : '서비스업',
        '유통업' : '운수 및 창고업',
        '운수창고업' : '운수 및 창고업',
        '전기가스업' : '에너지업',
        '통신업' : '정보통신업',
        '섬유의복' : '제조업',
        '음식료품' : '제조업',
        '기타제조업' : '제조업',
        '종이목재' : '제조업',
        '철강금속' : '제조업',
        '기계' : '제조업',
        '운수장비' : '제조업',
        '화학' : '제조업',
        '전기전자' : '제조업',
        '비금속광물' : '제조업',
        '의료정밀' : '제조업',
        '농업, 임업 및 어업' : '1차 산업',
        '광업' : '1차 산업',
        '의약품' : '의약품'
    })


    df_1 = df.set_index(['업종명', '종목명']).sort_index()

    return df_1
    
```
