```python
import pandas as pd
```

# 2011년

```python
df_2011=pd.read_csv('./OO_origin/Origin_11.csv',index_col='종목명')
df_list=df_2011.index.to_list()

### 연결 제무 제표, 제무제표 가져오기 --> 연결 제무 제표로 회사명(종목명)리스트 만들기
a=pd.read_csv('./연결기준data/2011data(연결).csv',encoding='euc-kr')
a['회사명']=a['회사명'].apply(lambda x: x.replace('(주)',''))

b=pd.read_csv('./개별기준data/2011data(개별).csv',encoding='euc-kr')
b['회사명']=b['회사명'].apply(lambda x: x.replace('(주)',''))

alist=a['회사명'].to_list()
blist=b['회사명'].to_list()

a.columns=['회사명','거래소코드','회계년도','총자산','부채비율','자기자본구성비율','매출액증가율','현금흐름','PBR최고','PBR최저','매출액','평균배당률']

b.columns=['회사명','거래소코드','회계년도','총자산','부채비율','자기자본구성비율','매출액증가율','현금흐름','PBR최고','PBR최저','매출액','평균배당률']


set1=set(df_list)
set2=set(alist)
set3=set(blist)
common_values = list(set1 & set2 & set3)
#len(common_values)
a=a.set_index('회사명').loc[common_values].sort_index()
b=b.set_index('회사명').loc[common_values].sort_index()
a=a.fillna(value=b)

df_2011=df_2011.loc[common_values]

N_2011=pd.concat([a,df_2011],axis=1)
N_2011['총자산']=N_2011['총자산']*1000
N_2011['현금흐름']=N_2011['현금흐름']*1000
N_2011['매출액']=N_2011['매출액']*1000
N_2011['현금흐름비율']=N_2011['현금흐름']/N_2011['매출액']
N_2011['PBR']=(N_2011['PBR최고']+N_2011['PBR최저']) / 2

N_2011=N_2011.drop(['거래소코드','회계년도','현금흐름','PBR최고','PBR최저','매출액'],axis=1)
N_2011.columns=['총자산규모(Size)','부채비율(DEB)','자기자본비율(DEB)','매출액증가율','배당률','ESG','E','S','G','평가연도','업종명','현금흐름(CF)','PBR']
N_2011

```
