```python
def change(x):
    if x==0:
        return '무배당기업'
    elif x>=47.12:
        return '고배당기업'
    else:
        return '일반배당기업'


train['배당률']=train['배당률'].apply(change)
```

```python
from scipy.stats import mstats
```

```python
dept = train['부채비율'].values
plt.boxplot(dept)
```

```python
dept = train['부채비율'].values
winsorized_dept = mstats.winsorize(dept, limits=[0.01, 0.01])
plt.boxplot(winsorized_dept)
train['부채비율']=winsorized_dept
```

```python
dept = train['자기자본비율'].values
winsorized_dept = mstats.winsorize(dept, limits=[0.01, 0])
plt.boxplot(winsorized_dept)
train['자기자본비율']=winsorized_dept
```

```python
dept = train['매출액증가율'].values
winsorized_dept = mstats.winsorize(dept, limits=[0, 0.01])
plt.boxplot(winsorized_dept)
train['매출액증가율']=winsorized_dept
```

```python
dept = train['현금흐름'].values
winsorized_dept = mstats.winsorize(dept, limits=[0.01, 0.01])
plt.boxplot(winsorized_dept)
train['현금흐름']=winsorized_dept
```

```python
dept = train['PBR'].values
winsorized_dept = mstats.winsorize(dept, limits=[0.01, 0.015])
plt.boxplot(winsorized_dept)
train['PBR']=winsorized_dept
```
