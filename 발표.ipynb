```python
import pandas as pd
import numpy as np
```

```python
a=pd.read_csv('./최종연도별data/Train.csv', index_col='Unnamed: 0')
a
```

```python
b=pd.read_csv('./최종연도별data/Test.csv', index_col='Unnamed: 0')
b
```

```python
c=pd.concat([a,b],axis=0)
c
```

```python
c.info()
```

```python
c.describe()
```

```python
c['ESG']=c['ESG'].map({
    'A+': 'A+',
    'A' : 'A',
    'B+': 'B+',
    'B' : 'B',
    'B이하' : 'B',
    'C':'C',
    '-': 'C',
    'D':'D'
})
c['E']=c['E'].map({
    'A+': 'A+',
    'A' : 'A',
    'B+': 'B+',
    'B' : 'B',
    'B이하' : 'B',
    'C':'C',
    '-': 'C',
    'D':'D'
})
c['S']=c['S'].map({
    'A+': 'A+',
    'A' : 'A',
    'B+': 'B+',
    'B' : 'B',
    'B이하' : 'B',
    'C':'C',
    '-': 'C',
    'D':'D'
})
c['G']=c['G'].map({
    'A+': 'A+',
    'A' : 'A',
    'B+': 'B+',
    'B' : 'B',
    'B이하' : 'B',
    'C':'C',
    '-': 'C',
    'D':'D'
})

```

```python
c['G'].value_counts()
```

```python
c.dropna(inplace=True)
```

```python
new_order =['업종명','평가연도','총자산규모(Size)','부채비율(DEB)','자기자본비율(DEB)','매출액증가율','현금흐름(CF)','PBR','ESG','E','S','G','배당률']
c = c.reindex(columns=new_order)
```

```python
c.info()
```

```python
def change(x):
    if x==0:
        return '무배당기업'
    elif x>=47.12:
        return '고배당기업'
    else:
        return '일반배당기업'


c['배당률']=c['배당률'].apply(change)
```

```python
c['배당기업구분'].value_counts()
```

```python
c=pd.get_dummies(c,columns=['ESG','E','S','G'])
```

```python
c = c[[col for col in c.columns if col != '배당률'] + ['배당률']]
```

```python
c.columns= ['업종명', '평가연도', '총자산규모(Size)', '부채비율(DEB)', '자기자본비율(DEB)', '매출액증가율',
       '현금흐름(CF)', 'PBR', 'ESG_A', 'ESG_A+', 'ESG_B', 'ESG_B+', 'ESG_C',
       'ESG_D', 'E_A', 'E_A+', 'E_B', 'E_B+', 'E_C', 'E_D', 'S_A', 'S_A+',
       'S_B', 'S_B+', 'S_C', 'S_D', 'G_A', 'G_A+', 'G_B', 'G_B+', 'G_C', 'G_D',
       '배당기업구분']
```

```python
c
```

```python
from sklearn.model_selection import train_test_split

train,test= train_test_split(c,test_size =0.2, shuffle=True,random_state=0)
```

```python
from sklearn.model_selection import train_test_split


df=pd.read_csv('./진짜최종데이터/real_sca_mm_train.csv',encoding='euc-kr',index_col='Unnamed: 0')
x=df.iloc[:,2:8].values
y=df['배당기업구분']
X_train, X_test, y_train, y_test = train_test_split(x,y, stratify=y, test_size=0.2, random_state=0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report

x_train, x_test, y_train, y_test = train_test_split(x,y, stratify=y, test_size=0.2, random_state=0)
tree = DecisionTreeClassifier(max_features=0.1,random_state=0,max_depth=30,min_samples_leaf=10)
tree.fit(x_train, y_train)
pred= tree.predict(x_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```
